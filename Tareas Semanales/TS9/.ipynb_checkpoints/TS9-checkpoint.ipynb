```python
import sympy as sp
import numpy as np
import scipy.signal as sig
import scipy.io as sio
import matplotlib.pyplot as plt
from pytc2.sistemas_lineales import  analyze_sys, plot_plantilla
import warnings
warnings.filterwarnings('ignore')
```

# Parte 1 - Diseño de filtros

Se necesita eliminar las interferencias de una señal electrocardiográfica (ECG) que fue registrada a una frecuencia de muestreo de 1 kHz. Para ello se propone la siguiente plantilla de diseño:


<img src='./Imagenes/Plantilla.png' style="height: 400px; width:auto;"></img>

y se pide diseñar dos filtros: uno IIR y otro FIR que cumplan con la plantilla, utilizando alguno de los métodos estudiados. Mostrar la respuesta en frecuencia, para ambos filtros, de:

- Módulo

- Fase

- Retardo

- Y respuesta al impulso.

## Diseño de Filtro IIR

```python
fs = 1000
nyq = fs/2

#------ Plantilla ------

wp = np.array([3, 25]) / nyq
ws = np.array([1, 35]) / nyq
gpass = 0.01
gstop = 40

#------ Diseño ------

num, den = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='butter', output='ba')
sos_iir = sig.iirdesign(wp, ws, gpass, gstop, analog=False, ftype='butter', output='SOS')

wrad, hh = sig.sosfreqz(sos_iir, worN=1000)
ww = wrad/np.pi * nyq

Wgd, gd = sig.group_delay((num, den))
wwgd = Wgd/np.pi * nyq

t = np.arange(0, 10, 0.01)
imp = np.fft.ifft(hh)

#------------------------------- Graficos -------------------------------

#------ Modulo ------
modulo = plt.figure()
plt.title('Filtro IIR: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.grid()
plt.axis([0,100,-50, 5])
#plt.show()

plot_plantilla(filter_type = 'bandpass', fpass = [wp[0]* nyq, wp[1]* nyq ], ripple = gpass , 
               fstop = [ws[0]* nyq, ws[1]* nyq ], attenuation = gstop, fs = fs)

#------ Fase ------
fase = plt.figure()
plt.title('Filtro IIR: Fase')
angles = np.unwrap(np.angle(hh))
plt.plot(ww, angles)
plt.ylabel('Fase [rads]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.axis([0,100,-50, 5])
plt.show()

#------ Retardo ------
ret = plt.figure()
plt.title('Filtro IIR: Retardo')
plt.plot(wwgd, gd)
plt.ylabel('Retardo [s]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.show()

#------ Respuesta al impulso ------
impulso = plt.figure()
plt.title('Filtro IIR: Respuesta al Impulso')
plt.plot(t, imp)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo [s]')
plt.grid()
plt.show()
```

## Diseño de Filtro FIR

```python
fs = 1000
nyq = fs/2

den_f = 1.0
cant_coef = 3001

#------ Plantilla ------

fpass = np.array([1, 35])
fstop = np.array([3, 25])
gpass = 0.01
gstop = 40 

frecs = [0.0,       fpass[0],     fstop[0],     fstop[1],     fpass[1],     nyq]
gains = [-gstop,   -gstop,       -gpass,         -gpass,      -gstop,      -gstop] 
gains = 10**(np.array(gains)/20)

#------ Disenio ------

fir_win = sig.firwin2(cant_coef, frecs, gains , window='blackmanharris', fs=fs )

wrad, hh = sig.freqz(fir_win, den_f, worN = 1000)
ww = wrad/np.pi * nyq

gd = -np.diff(np.unwrap(np.angle(hh)))/np.diff(ww)

t = np.arange(0, 10, 0.01)
imp = np.fft.ifft(hh)

#------------------------------- Graficos -------------------------------

#------ Modulo ------
modulo = plt.figure()
plt.title('Filtro FIR: Modulo')
plt.plot(ww, 20 * np.log10(abs(hh)))
plt.xlabel('Frecuencia [Hz]')
plt.ylabel('Modulo [dB]')
plt.axis([0,100,-50, 5])
plt.grid()
#plt.show()

plot_plantilla(filter_type = 'bandpass', fpass = [frecs[2], frecs[3]], ripple = gpass , fstop = [ frecs[1], frecs[4] ], 
               attenuation = gstop, fs = fs)

#------ Fase ------
fase = plt.figure()
plt.title('Filtro FIR: Fase')
angles = np.unwrap(np.angle(hh))
plt.plot(ww, angles)
plt.ylabel('Fase [rads]')
plt.xlabel('Frecuencia [Hz]')
#plt.axis([0,100,0, 100])
plt.grid()
plt.show()

#------ Retardo ------
ret = plt.figure()
plt.title('Filtro FIR: Retardo')
plt.plot(ww[1:], gd)
plt.ylabel('Retardo [s]')
plt.xlabel('Frecuencia [Hz]')
plt.grid()
plt.axis([0,100,-4, 0])
plt.show()

#------ Respuesta al impulso ------
impulso = plt.figure()
plt.title('Filtro FIR: Respuesta al Impulso')
plt.plot(t, imp)
plt.ylabel('Amplitud')
plt.xlabel('Tiempo [s]')
plt.grid()
plt.show()
```

# Parte II - Implementación y uso

En esta parte utilizaremos los dos filtros diseñados para mitigar las interferencias.

### Cargamos los datos:

```python
mat_struct = sio.loadmat('ecg.mat')
ecg_one_lead = mat_struct['ecg_lead']
ecg_one_lead = ecg_one_lead.flatten()
cant_muestras = len(ecg_one_lead)

fig_sz_x = 10
fig_sz_y = 7
fig_dpi = 100 # dpi

regs_interes = ( 
        np.array([5, 5.2]) *60*fs, # minutos a muestras
        np.array([12, 12.4]) *60*fs, # minutos a muestras
        np.array([15, 15.2]) *60*fs, # minutos a muestras
        [4000, 5500], # muestras
        [10e3, 11e3], # muestras
)
```

## Filtrado con Filtro IIR

```python
ECG_f_iir = sig.sosfilt(sos_iir, ecg_one_lead)

demora_iir = 115

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)
    plt.plot(zoom_region, ECG_f_iir[zoom_region + demora_iir], label='Filtrado IIR')
    #plt.plot(zoom_region, ECG_f_win[zoom_region + demora], label='Win')
    
    plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```

Se observa que en las bajas frecuencias realiza un buen filtrado, pero despues introduce una alta distorsión a frecuencias más altas.

Realizamos un filtrado bidireccional para comprobar si es ocasionada por una misma distorsion de fase del filtro.

## Filtrado Bidireccional con IIR

```python
ECG_f_iir_b = sig.sosfiltfilt(sos_iir, ecg_one_lead)

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)
    plt.plot(zoom_region, ECG_f_iir_b[zoom_region ], label='Filtrado IIR Bidireccional')
    #plt.plot(zoom_region, ECG_f_win[zoom_region + demora], label='Win')
    
    plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```

Podemos ver que el filtrado bidireccional logró un mejor filtrado sin distorsión, por lo cual pudimos comprobar que lo anterior era ocasionado por una distorsión de fase.


## Filtrado con FIR

```python
ECG_f_win = sig.lfilter(fir_win, den_f, ecg_one_lead)

demora_fir = 1500

for ii in regs_interes:
    
    # intervalo limitado de 0 a cant_muestras
    zoom_region = np.arange(np.max([0, ii[0]]), np.min([cant_muestras, ii[1]]), dtype='uint')
    
    plt.figure(figsize=(fig_sz_x, fig_sz_y), dpi= fig_dpi, facecolor='w', edgecolor='k')
    plt.plot(zoom_region, ecg_one_lead[zoom_region], label='ECG', linewidth=2)
    #plt.plot(zoom_region, ECG_f_iir_b[zoom_region + demora_iir], label='Filtrado IIR Bidireccional')
    plt.plot(zoom_region, ECG_f_win[zoom_region + demora_fir], label='Filtrado FIR')
    
    plt.title('ECG filtering example from ' + str(ii[0]) + ' to ' + str(ii[1]) )
    plt.ylabel('Adimensional')
    plt.xlabel('Muestras (#)')
    
    axes_hdl = plt.gca()
    axes_hdl.legend()
    axes_hdl.set_yticks(())
            
    plt.show()
```

Podemos ver que en este caso filtra bien el FIR, y no tiene distorsión de fase, esto debido a que su fase es lineal.

Por eso no es necesario realizar un filtrado bidireccional.
